# JSON

### So What is JSON?

****JSON stands for Java-Script Object Notation****

****JSON is a format used to share, exchange and store data****

****Almost all the API's (Application Program interfcaes) and databases use this format to exchange information****

****It is readable by machines and humans alike****

### JSON Format

****JSON format is simple and easily readable for both computers and machines****

****It's similar to Python's dictionary, which we have already covered****

****Key : Value pairs****

#### Look Closely! the json is similar to Python's dictionary, You may notice a couple of nested lists and dictionaries but we all know how to handle nested lists and dictionaries, right?

### JSON and Python

****Python comes with built-in package (collection of modules) to handle JSON data****

****Build-in package: "json"****

### 1) Serialization (Data to JSON Formatting)

****The process of converting data into JSON format is known as Serialization****

****The data is encoded in order to share/exchange this information across different networks****

```python
from json import dumps
```

****dumps() function from Pythons's package is used to encode data into JSON format****

****dump implies putting all the data into this function in order to convert it into JSON****

****The 's' in the dumps() function indicates that the input data will be in 'string' format****

****dumps() allows us to convert Python objects into json objects****

```python
data = {'Name': 'Jehad', 'age': 23, 'Favorite food': 'Kachalon'}
print(f'{data} is an example of data that we will be converting into Json => {type(data)}')
```

```python
encoding = dumps(data)
print(f'{encoding} is an example of data which is encoded as Json => {type(encoding)}')
```

### 2) List to JSON

```python
myList = [1,2,3.3,4.3,'this is an example of converting list into JSON']
print(myList, type(myList))

l_to_J = dumps(myList)
print(l_to_J, type(l_to_J))
```

```python
colors = ['Red', 'Green', 'Orange', 'Yellow', 'Purple', 'White', 'Violet', 'Golden', 'Silver', 'Blue']
print(colors, type(colors))

Jcolor = dumps(colors)
print(Jcolor, type(Jcolor))
```

### 3) Tuple to JSON

### 4) String to JSON 

```python
string = 'Kyun kia SamjhaY??'
print(string, type(string))

Jstring  = dumps(string)
print(Jstring, type(Jstring))
```

### 5) Integer to JSON

```python
integer = 152
print(integer, type(integer))

Jinteger = dumps(integer)
print(Jinteger, type(Jinteger))
```

### 6) Float to JSON

```python
floating = 152.0315
print(floating, type(floating))

Jfloat = dumps(floating)
print(Jfloat, type(Jfloat))
```

```python
print(True, type(True))

Jtrue = dumps(True)
print(Jtrue, type(Jtrue))

print(False, type(False))

Jfalse = dumps(False)
print(Jfalse, type(Jfalse))
```
