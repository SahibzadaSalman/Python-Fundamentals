### All Parameters in One Functions?

****Can we add all type of arguments in one functions?****

#### Yes We Can!!!!

#### But HOWWWW!!!!!

```python
def all_in_one(x , y, *args, **kwargs):
    print('This is an example of using all type of arguments in one function')
    print(F'x = {x}', type(x))
    print(F'y = {y}', type(y))
    print(F'args = {args}', type(args), len(args))
    print(F'kwargs = {kwargs}', type(kwargs), len(kwargs))
```

### Recursion

****The process in which a function calls itself directly or indirectly is known as recursion****

****In Python We can call functions in other functions****

****In recursion, instead of calling other functions, the function calls iteself :p****

****The function that calls itself is known as recursive function****

##### Example: Writing a program that calculates sum of all the numbers from 1 to a given number

```python
def addition(arg):
    if arg > 0:
        add = arg + addition(arg-1)
        return add 
    return 0
```

```python
addition(10)
```

##### Example: Writing a function to calculate factorial of a given number

```python
def factorial(arg):
    """This function is used to find the factorial of a given number"""
    if arg == 1 or arg == 0:
        return 1
    else:
        print(arg)
        fact = arg * factorial(arg-1)
        return fact
```

```python
num = int(input('Enter a Number: '))
print(F'The factorial of {num} is: {factorial(num)}')
```

```python
factorial(3)
```

```python
factorial(4)
```

```python
factorial(5)
```

****In above example factorial is a recursive function as it calls itself****

****Recursive function is like a while loop it runs untill a certain condition is TRUE****

```python
def recursion(arg):
    if arg > 0:
        result = arg + recursion(arg - 1)
        print(result)
    else:
        result = 0
    return result
```

```python
recursion(6)
```
