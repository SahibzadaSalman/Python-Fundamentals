## 1) if Statement

### Syntax

if condition:
    perform task

****input() is a Python's built-in function that takes the input from user****

```python
x = input("Enter 1st Value: ")
x = int(x)
y = input("Enter 2nd Value: ")
y = int(y)

if x == y:
    print('x is equal to y')
```

****In the above example, " x == y " is the condition****

****The body of " if " is executed only if the above condition is met****

****When the variable x is equal to variable y, the given condition is true and statements inside the body of " if " are executed****

****If x is not equal to y , the given condition is false and instructions inside the body of " if " are ignored****

****The print() function lies outside of the if statement therefore, it is executed regardless of the condition****

****The statement only runs when the given condition is met otherwise it won't run****

```python
a = 3
b = 5
if a < b
    print('a is greater than b')
```

## 2) if-else Statement

****if the condition in true then " if " body is excuted****
****otherwise else body is executed****

```python
a = int(input("Enter a number: "))

if a%2 == 0:
    print('The number', a, 'is an Even number')
else:
    print('The number', a, 'is an Odd number')
```

****Explaining the above example:****

****input() function first takes the input number from user*****

****The (a % 2 == 0) is the condition which means the "remainder of a divided by 2 is equal to 0" then****

* ****if this condition is True then run the body of " if " and everything inside else body is ignored/skipped****

* ****if the given condition is not met then run the body of else statement and ignore the " if " body****


```python
x = int(input('Enter the value of x: '))
y = int(input('Enter the value of y: '))

if x > y:
    print('\nx is greater than y\n')
    print('(Running if block)\n')
else:
    print('\ny is greater than x\n')
    print('(Running else block)')
```

## 3) if-elif-else Statements

#### In if-elif-else statements;
#### 1) multiple conditions are checked 
#### 2) if the condition for " if " is FALSE, then the condition in " elif " is executed and so on 
#### 3) if all the conditions are found to be FALSE then and only then the body of " else " is executed

```python
a = int(input('Enter a value of a: '))
b = int(input('Enter a value of b: '))

if a > b: 
    print('a is greater than b')
elif a < b:
    print('a is smaller than b')
else:
    print('a and b are equal')
```

```python
a = int(input('Enter a value of a: '))

if a > 0:
    print('The given number is Positive')
    if a%2 == 0:
        print(a, 'is an Even number')
    else:
        print(a, 'is an Odd number')
elif a < 0:
    print('The given number is Negative')
    if a%2 == 0:
        print(a, 'is a negative even')
else:
    print('The given number is Zero')
```

## Conditional Expressions OR Python's Ternary Operator

****Operators that evaluate something based on condition being true or false****

****The condition is first executed, and if turns out to be "True" than first statement is executed otherwise second statement is executed****

#### Syntax

****1) Condition is evaluated first****

****2) If "condition" holds "True" then Statement No 1 is executed****

****3) If "condition" holds "False" then Statement No 2 is executed****

```python
Weather = input('What is the weather today? ')

print("Don't leave the house today!" if Weather == 'raining' else "You can go wherever you want to!" )
```

```python
a = int(input('Enter a number: '))
b = int(input('Enter a 2nd number: '))

m = (a + 5) if a > b else (b + 5)
print(m)
```

```python
x = int(input('Enter the value of x: '))
y = int(input('Enter the value of y: '))

z = (x-2)  if x >= y else (y+2)
print(z)
```

****Conditional statements can also be combined into multiple conditions like alternative if-elif-else statements****

```python
x = int(input('Enter a number: '))
series = ('Peshawar' if x <= 10 else
        'Islamabad' if (x > 10 and x <= 20) else
         'Karachi' if (x > 20 and x <= 30) else
         'Lahore' if (x > 30 and x <= 40) else
         'DI Khan' if (x >= 40) else 'Enter a correct number'
         )
print(series)
```
