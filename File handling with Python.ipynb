### Files

****Files are used to permanently stored information/data in computers (hard drives)****

****Python allows us to create, store and manipulate data in files****

### File Operations

****While working with files in Python, we need to follow 3 steps****

****1) Open a file****

****2) Perform operations in the file (reading/writing)****

****3) Close the file****

### 3) Open a FIle

****Python has built-in funciton called "open()"****

****open() usually takes 2 parameters, 1st one is the file name and the second one shows the 'mode' in which we're opening a file (either read, write or other)****

### Syntax

```python
file = open('sample.txt')
```

#### Modes

##### There different modes in which we can open our file

****r ===> reading a file (default)****

****w ===> wirting something in the file****

****a ===> appending something in the file****

****b ===> opening in binary****

****r+ ===> both reading and writing but this requires the file to be present in the directory****

****w+ ===> both reading and writing, will create the file even if not present****

#### 1) close() function

```python
file = open('sample.txt','r')
print(file.read())
```

#### 2) try-finally block

```python
try:
    file = open('sample.txt', 'r')
    print(file.read())
finally:
    file.close()
```

****Now if we try to read the file, it'll throw an error!****

```python
print(file.read())
```

****Try-finally block may not completely run if we have an exception in the code, also it is not always guarenteed that the file will be closed****

#### 3) with statement

```python
with open('sample.txt', 'r') as file:
    print(file.read())
```

****with statement make sures to close the file as soon as the code inside the with statement is executed****

****Note! we don't need to explicitly call the close() function to close the file****

#### readline()

****readline() function is used to print the file line by line****

```python
new = open('Tariqullah.txt', 'r')
print(new.readline())   # each readline() function prints exactly one line from the file
print(new.readline())   # the number of lines printed will depend on how many times the readline() is called
print(new.readline())
```

#### We can also use for loop to print all the lines in the file

```python
new = open('Tariqullah.txt', 'r')
```

```python
for line in new:
    print(line)
```

#### readlines()

****readlines() function returns all the lines in the file****

```python
new = open('Tariqullah.txt', 'r')
print(new.readlines())
```

****Let's add sequences in our file****

#### Appending dictionary

```python
dictt = {'key1': 'Adding', 'key2' : 'Dictionary', 'key3': 'in', 'key4': 'our', 'key5': 'file'}
print(dictt)

with open('sample.txt', 'a') as file:
    file.write('Adding sequences in our file: \n')
    file.write(f'\t\tdictionary: {str(dictt)}\n')
```

```python
file = open('sample.txt', 'r+')
print(file.read())
```

#### Appending a list

```python
color = ['Red', 'Blue', 'Green', 'Yellow', 'Orange']
print(color)
with open('sample.txt', 'a') as file:
    file.write(f'\t\tList: {str(color)}\n')
```

```python
file = open('sample.txt', 'r+')
print(file.read())
```

#### Appending a Tuple

```python
fruits = ['Apple', 'Berry', 'Guava', 'Mango', 'Orange']
print(fruits)
with open('sample.txt', 'a') as file:
    file.write(f'\t\tTuple: {str(fruits)}\n')
```

```python
file = open('sample.txt', 'r+')
print(file.read())
```

```python
file.close()
```

#### Splitting lines into individual words

```python
with open('sample.txt', 'r+') as file:
    print(file.read())
```

#### splitting all the lines in our file into words using split() function

```python
with open('sample.txt', 'r+') as file:
    for line in file:
        print(line.split())
        #print('\n',word)
```

#### printing number of words in each line

```python
with open('sample.txt', 'r+') as file:
    for line in file:
        print(line.split(), 'No. of words =====>', len(line.split()))
        #print('\n',word)
```
